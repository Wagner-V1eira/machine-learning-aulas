## Tarefa 1: Carregamento e Exploração dos Dados

Carregue o dataset MNIST e explore suas características.

**O que fazer:**

- Carregue o dataset MNIST usando `fetch_openml('mnist_784', version=1, parser='auto')`
- Visualize algumas imagens do dataset
- Analise a distribuição das classes
- Verifique as dimensões dos dados


```python
# Carregando o dataset MNIST
print("Carregando dataset MNIST...")
mnist = fetch_openml("mnist_784", version=1, parser="auto")

X = mnist.data.values if hasattr(mnist.data, "values") else mnist.data
y = mnist.target.values if hasattr(mnist.target, "values") else mnist.target
y = y.astype(int)

print(f"Shape de X: {X.shape}")
print(f"Shape de y: {y.shape}")
print(f"Valores únicos em y: {np.unique(y)}")

# Visualizar algumas imagens
fig, axes = plt.subplots(2, 5, figsize=(12, 6))
for i, ax in enumerate(axes.flat):
    ax.imshow(X[i].reshape(28, 28), cmap="gray")
    ax.set_title(f"Dígito: {y[i]}")
    ax.axis("off")
plt.suptitle("Exemplos do Dataset MNIST")
plt.tight_layout()
plt.show()

# Distribuição das classes
classes, counts = np.unique(y, return_counts=True)
plt.figure(figsize=(10, 5))
plt.bar(classes, counts)
plt.xlabel("Dígito")
plt.ylabel("Quantidade")
plt.title("Distribuição das Classes no MNIST")
plt.xticks(classes)
plt.grid(axis="y", alpha=0.3)
plt.show()

print(f"\nDistribuição das classes:")
for digit, count in zip(classes, counts):
    print(f"Dígito {digit}: {count} amostras ({count/len(y)*100:.2f}%)")
```
